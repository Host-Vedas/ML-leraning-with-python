```python
import pandas as pd
import random
```

```python
L = []
```

```python
for i in range(1000):
    L.append(random.randint(1,6))
```

```python
len(L)
```

```python
L[:5]
```

```python
s = (pd.Series(L).value_counts()/pd.Series(L).value_counts().sum()).sort_index()
```

```python
s
```

```python
s.plot(kind='bar')
```

```python
L = []
for i in range(1000):
    a = random.randint(1,6)
    b = random.randint(1,6)
    L.append( a + b)
```

```python
len(L)
```

```python
L[:5]
```

```python
s1 = (pd.Series(L).value_counts()/pd.Series(L).value_counts().sum()).sort_index()
```

```python
s1.plot(kind='bar')
```
